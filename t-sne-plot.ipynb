```python
import pandas as pd

a=pd.read_csv('output.model.vec', sep=' ', skiprows=[0])
```

```python
import sklearn.manifold
import numpy as np

df = a
N = 10000
rndperm = np.random.permutation(df.shape[0])
df_subset = df.loc[rndperm[:N],:].copy()
data_subset = df_subset.values
TSNE = sklearn.manifold.TSNE

tsne = TSNE(n_components=2, verbose=1, perplexity=20, n_iter=300)
tsne_results = tsne.fit_transform(data_subset[:,1:-1])
```

```python
# Basic
import numpy as np
import pandas as pd

# Plots
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.preprocessing import StandardScaler

# Clustering
from sklearn import cluster

# Bokeh
from bokeh.io import output_notebook
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.models import HoverTool
%matplotlib inline
output_notebook()

N = 10000
labels = data_subset[:N,0]
x = []
y = []
for value in tsne_results[0:N]:
    x.append(value[0])
    y.append(value[1])

    

source_train = ColumnDataSource(
        data=dict(
            x = tsne_results[:len(y),0],
            y = tsne_results[:len(y),1],
            desc = labels,
            colors = ["#%02x%02x%02x" % (int(r), int(g), int(b)) for r, g, b, _ in 
                      255*mpl.cm.jet(mpl.colors.Normalize()(y))],
        )
    )


hover_tsne = HoverTool(names=["test", "train"], tooltips=[("Name", "@desc"), ])
tools_tsne = [hover_tsne, 'pan', 'wheel_zoom', 'reset']
plot_tsne = figure(plot_width=1000, plot_height=600, tools=tools_tsne, title='Bacterias')

plot_tsne.circle('x', 'y', size=10, fill_color='colors', 
                 alpha=0.5, line_width=0, source=source_train, name="train")

show(plot_tsne)
```
